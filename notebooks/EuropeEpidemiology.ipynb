## Compare Confirmed data between JHU and OWID  

```python
confirmed_owid_tests=owid.get_stats(which='total_deaths',location=loc)
confirmed_owid_tests=confirmed_owid_tests.set_index(['location','date'])
confirmed_jhu_tests=jhu.get_stats(which='deaths',location=loc)
confirmed_jhu_tests=confirmed_jhu_tests.set_index(['location','date'])

result = pd.concat([confirmed_owid_tests,confirmed_jhu_tests],axis=1, sort=False)
result['ratio'] = result['total_deaths']/result['deaths']
result=result.reset_index()
mask = (result['date'] > '2020-07-01')
result = result.loc[mask] 
result = result.set_index("date")

fig=coco.cocoa_basic_plot(result,'ratio')
show(fig)
```
