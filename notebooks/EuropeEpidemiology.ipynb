```python
import pandas as pd
from bokeh.io import show, output_notebook
from bokeh.layouts import column,row
output_notebook(hide_banner=True)
from cocoa import covid19 as co
from cocoa import display as cd
from cocoa import world as cw
import importlib
w=cw.WorldInfo()
```

## Europe with JHU

```python
jhu = co.DataBase("jhu")
coco = cd.CocoDisplay()
```

```python
loc= ['France','Spain','Italy','Greece','Belgium','Germany','Austria','Israel']
deaths_jhu=jhu.get_stats(which='deaths',location=loc)
fig=coco.cocoa_basic_plot([deaths_jhu],['deaths'])
show(fig)
```

compare "deaths" for Spain in the case of option = nonneg and without

```python
confirmed_jhu_ne=jhu.get_stats(which='deaths',location='Spain')
confirmed_jhu=jhu.get_stats(which='deaths',location='Spain',option='nonneg')
print(confirmed_jhu)
fig=coco.cocoa_basic_plot([confirmed_jhu_ne,confirmed_jhu],['deaths'])
show(fig)
```

## Europe with OWID

```python
owid =  co.DataBase('owid')
```

```python
loc= ['France','Spain','Italy','Greece','Belgium','Germany','Austria','Israel']
owid_deaths=owid.get_stats(which='total_deaths',location=loc)
fig=coco.cocoa_basic_plot([owid_deaths],['total_deaths'])
show(fig)
```

```python
owid_deaths=owid.get_stats(which='total_deaths',location='France')
jhu_deaths=jhu.get_stats(which='deaths',location='France')

result = pd.concat([owid_deaths,jhu_deaths],axis=1, sort=False)
result['ratio']=result['total_deaths']/result['deaths']
fig=coco.cocoa_basic_plot([result],['ratio'])
show(fig)
```

## Compare Confirmed data between JHU and OWID  

```python
confirmed_owid_tests=owid.get_stats(which='total_cases',location=loc)
confirmed_owid_tests=confirmed_owid_tests.set_index(['location','date'])
confirmed_jhu_tests=jhu.get_stats(which='confirmed',location=loc)
confirmed_jhu_tests=confirmed_jhu_tests.set_index(['location','date'])

result = pd.concat([confirmed_owid_tests,confirmed_jhu_tests],axis=1, sort=False)
result['ratio'] = result['total_cases']/result['confirmed']
result=result.reset_index()
mask = (result['date'] > '2020-07-01')
result = result.loc[mask] 
result = result.set_index("date")

fig=coco.cocoa_basic_plot([result],['ratio'])
show(fig)
```

## Plot Active cases for JHU database 

```python
a=jhu.get_stats(location=loc,which='confirmed').reset_index()
b=jhu.get_stats(location=loc,which='recovered').reset_index()
c=jhu.get_stats(location=loc,which='deaths').reset_index()
```

```python
a['actives']=a['confirmed']-b['recovered']-c['deaths']
print(a)
fig = coco.cocoa_basic_plot([a],['actives'])
show(fig)
```
