We provide the backend python package cocoa which is available in GitHub : https://github.com/tjbtjbtjb/CoCoA
It allows the final user to use as simple methods as we can to get data, analyse and/or plot them. 

```python
from bokeh.io import show, output_notebook
output_notebook(hide_banner=True)
from cocoa import covid19 as co
from cocoa import display as cd
from cocoa import world as cw
```

# JHU's database

```python
jhu = co.DataBase('jhu')
coco = cd.CocoDisplay()
#countries= ['France','Spain','Italy','Greece','Germany','Belgium','United Kingdom','Israel']
stat_jhu = jhu.get_stats(which='confirmed',location='France')
confirmed_owid=owid.get_stats(which='new_cases',location='France')
confirmed_owid['diff']=confirmed_owid['new_cases'].shift(-1)
show(coco.cocoa_basic_plot([stat_jhu,confirmed_owid],['diff']))
```

# OWID's database

```python
owid =  co.DataBase('owid')
```

```python
countries= ['France','Spain','Italy','Greece','Germany','Belgium','United Kingdom','Israel']
confirmed_owid=owid.get_stats(which='new_cases_per_million',location=countries)
confirmed_owid['Sum7D'] = confirmed_owid['new_cases_per_million'].rolling(7).mean() 
fig=coco.cocoa_basic_plot([confirmed_owid],['new_cases_per_million'])
show(fig)
```
