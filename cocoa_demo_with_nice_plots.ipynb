Introduction

As physicists working at CERN experiments, used to dealing with big data, we wish to share our knowledge in statistics to the greatest number of people for Covid-19 analysis. Data mining and statistics should help people learn about the most important subject of current history. For this purpose we have designed the COvid COlab Analysis projet, aka COCOA. COCOA will provide open source online statistical tools with a simple user interface and simple modelization schemes. COCOA is based on the Google Colab infrastructure which supplies a popular Python Notebook framework. This gives the opportunity to have multi-user code editing and huge computing infrastructure (included GPU for future Deep Learning Covid-19 analysis). Moreover an original analysis based on particle physics analysis is under development and should be available soon in COCOA.

# Let's Draw 1 : plot 2 differents countries (interactive selection)

```
# Need the use of version >(stricly) 0.1
!pip install git+https://github.com/tjbtjbtjb/CoCoA.git
```

```
import cocoa.covid19 as cc
import cocoa.world as cw
from cocoaplot import display as cd

from bokeh.io import show, output_notebook
```

```
p=cc.Parser()

w=cw.WorldInfo()
CountriesEU=w.getEuropeanCountries()

coco = cd.CocoDisplay(p)

### Select data to play with
type_data='deaths'
cumul_or_diff='cumul'
country=CountriesEU

####
output_notebook()
myfig = coco.DefFigInteractive(type_data,cumul_or_diff,country)
show(myfig)
```

# Let's Draw 2: plot list of country (static selection)

```
coco = cd.CocoDisplay(p)
### Select data to play with
type_data='deaths'
cumul_or_diff='cumul'
#countries=['France',"China","US"]
countries=['France','Spain','Italy','Greece']

####
statfig=coco.DefFigStatic(type_data,cumul_or_diff,countries)
show(statfig)
```
