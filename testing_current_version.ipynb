```python
from cocoaplot import display as cd
from bokeh.io import show, output_notebook
```

```python
output_notebook()
CountriesEU=w.getEUCountries()
coco = cd.CocoDisplay()

### Select data to play with
type_data='confirmed'
cumul_or_diff='Diff'
countries=CountriesEU

myfig = coco.DefFigInteractive(which=type_data,type=cumul_or_diff,country=countries)
show(myfig)
```
